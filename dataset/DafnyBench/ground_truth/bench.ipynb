```python
# Other_Code dir 1
import os

def find_missing_files(original_dir, correct_dir, fail_dir):
    try:
        # Get list of files in each directory
        original_files = set(os.listdir(original_dir))
        correct_files = set(os.listdir(correct_dir)) if os.path.exists(correct_dir) else set()
        fail_files = set(os.listdir(fail_dir)) if os.path.exists(fail_dir) else set()

        # Find files that are in original but not in correct or fail
        missing_files = original_files - (correct_files | fail_files)

        return missing_files

    except FileNotFoundError as e:
        print(f"Error: {e}")
        return None

def write_missing_files_to_file(missing_files, output_file):
    try:
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write("Files in 'original' but not in 'correct' or 'fail':\n")
            for file_name in missing_files:
                file.write(file_name + '\n')
        
        print(f"Output written to {output_file}")

    except Exception as e:
        print(f"Error writing to file: {e}")

# Example usage with your provided directory paths and output file path
original_dir = r'Original_Code'
correct_dir = r'Correct_Code'
fail_dir = r'Fail_Code\Code'
output_file = r'output_missing_files.txt'

#Other_Code files
missing_files = find_missing_files(original_dir, correct_dir, fail_dir)

if missing_files:
    write_missing_files_to_file(missing_files, output_file)
else:
    print("Failed to find missing files.")
```

```python
# Other_Code dir 2
import shutil

def copy_missing_files_to_directory(missing_files, source_dir, destination_dir):
    try:
        # Ensure destination directory exists
        os.makedirs(destination_dir, exist_ok=True)

        for file_name in missing_files:
            source_path = os.path.join(source_dir, file_name)
            destination_path = os.path.join(destination_dir, file_name)
            shutil.copy(source_path, destination_path)
            print(f"Copied '{file_name}' to '{destination_dir}'")

    except Exception as e:
        print(f"Error copying files: {e}")


destination_dir = r'\Other_Code'
if missing_files:
    copy_missing_files_to_directory(missing_files, original_dir, destination_dir)
else:
    print("No missing files found.")
```

```python
# Fail_Bugs_Code Dir

import os
import shutil

# Define directories
correct_dir = r'Correct_Used_Code'
bugs_dir = r'All_Bugs_Code\Mutations'

# Function to get base filename without extension
def get_base_filename(filename):
    return os.path.splitext(filename)[0]

# Get list of files in correct directory
correct_files = os.listdir(correct_dir)

# Check for missing files in bugs directory
missing_files = []

for correct_file in correct_files:
    base_name = get_base_filename(correct_file)
    found = False
    i = 0
    while True:
        bugs_file = f"{base_name}_{i}.dfy"
        if bugs_file in os.listdir(bugs_dir):
            found = True
            break
        i += 1
        if i > 100:  # Assume no more than 100 variations (abc_1.dfy to abc_100.dfy)
            break
    
    if not found:
        missing_files.append(correct_file)

fail_bugs_dir = r'Fail_Bugs_Code'

if missing_files:
    print("Copying missing files to Fail_Bugs_Code directory:")
    for file in missing_files:
        src_file = os.path.join(correct_dir, file)
        dst_file = os.path.join(fail_bugs_dir, file)
        shutil.copy(src_file, dst_file)
        print(f"Copied {file} to {fail_bugs_dir}")
else:
    print("No missing files found.")
```

```python
import os

def get_filenames_without_extension(directory):
    """Returns a set of filenames without extensions in the given directory."""
    filenames = set()
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            name, _ = os.path.splitext(filename)
            filenames.add(name)
    return filenames

def main():
    dir1 = r"All_Bugs_Code\FaultLocalization"
    dir2 = r"All_Bugs_Code\Code"
    
    # Get filenames without extensions from both directories
    files_dir1 = get_filenames_without_extension(dir1)
    files_dir2 = get_filenames_without_extension(dir2)
    
    # Find files that are in dir1 but not in dir2
    unique_to_dir1 = files_dir2 - files_dir1
    
    # Write the result to output.txt with UTF-8 encoding
    with open('output.txt', 'w', encoding='utf-8') as file:
        if unique_to_dir1:
            file.write("Files in directory 1 but not in directory 2 (ignoring extensions):\n")
            for filename in unique_to_dir1:
                file.write(f"{filename}\n")
        else:
            file.write("No unique files found in directory 1.\n")

if __name__ == "__main__":
    main()
```

```python
import os
import shutil

def get_filenames_without_extension(directory):
    """Returns a set of filenames without extensions in the given directory."""
    filenames = set()
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            name, _ = os.path.splitext(filename)
            filenames.add(name)
    return filenames

def copy_unique_files_from_dir2(src_dir, dest_dir, unique_files, extension=".dfy"):
    """Copy files from src_dir to dest_dir that match the names in the unique_files set with the given extension."""
    for filename in unique_files:
        src_file_path = os.path.join(src_dir, f"{filename}{extension}")
        if os.path.exists(src_file_path):  # Ensure the file exists before copying
            dest_file_path = os.path.join(dest_dir, f"{filename}{extension}")
            shutil.copy(src_file_path, dest_file_path)

def main():
    dir1 = r"All_Bugs_Code\FaultLocalization"
    dir2 = r"All_Bugs_Code\Code"
    destination_dir = r"Extra_Fault_Loc_Code"
    
    # Get filenames without extensions from both directories
    files_dir1 = get_filenames_without_extension(dir1)
    files_dir2 = get_filenames_without_extension(dir2)
    
    # Find files that are in dir1 but not in dir2
    unique_to_dir1 = files_dir2 - files_dir1
    
    # Copy the matching .dfy files from dir2 to the destination directory
    copy_unique_files_from_dir2(dir2, destination_dir, unique_to_dir1)

if __name__ == "__main__":
    main()
```
